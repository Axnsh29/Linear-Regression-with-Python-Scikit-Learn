```python
#Importing all libraries required in this notebook

import pandas as pd #pandas is used for data analysis
import numpy as np  #numpy is used for numerical python
import matplotlib.pyplot as plt  #matplotlib is used for visualization
%matplotlib inline
from sklearn.model_selection import train_test_split #sklearn is efficient tool for predictive data analysis
from sklearn.linear_model import LinearRegression  
from sklearn import metrics  
```

```python
#reading data from the remote link

url = "http://bit.ly/w-data"
student_data = pd.read_csv(url)
print("Student data imported successfully")

student_data.info()
```

```python
#Plotting the distribution of scores
student_data.plot(x='Hours', y='Scores' , style = '*' , color = 'r')  
plt.title('Student Data - Hours vs Percentage')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score')  
plt.show()
```

```python
#dividing the data into attributes and labels
X = s_data.iloc[:, :-1].values  
Y = s_data.iloc[:, 1].values

print(f"Attributes values are \n {X}")
print(f"Label values are \n {Y}")
```

```python
#spliting the data into training and test sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
#fitting training data into model
model = LinearRegression()  
model.fit(X_train, Y_train) 
print("Training complete.")
```

```python
#plotting the regression line
line = model.coef_*X+model.intercept_
```

```python
#plotting for the test data
plt.scatter(X, Y, color = 'r')
plt.plot(X, line );
plt.show()
```

```python
#making predictions
print(X_test) 
Y_pred = model.predict(X_test) 
```

```python
#comparing Actual data vs Predicted data
actual = Y_test
predicted = Y_pred
df = pd.DataFrame({'Actual': actual, 'Predicted': predicted})
print(df)
```

```python
#what will be the score if student studies for 9.25 hrs/day?
try:
    hours = 9.25
    own_pred = model.predict([[hours]])
    print("No of Hours = {}".format(hours))
    print("Predicted Score = {}".format(own_pred[0]))
except Exception as e:
    print("An error occurred:", str(e))
```

```python
#Evaluating the model
# Assuming you have y_test and y_pred
Y_true = Y_test
Y_predicted = Y_pred

# Calculate Mean Absolute Error (MAE)
mae = metrics.mean_absolute_error(Y_true, Y_predicted)

# Optionally, you can print and customize the output
print('Mean Absolute Error:', mae)
```
